We check again how close the data collected on the physical car resembles the data collected on the simulator on the same map.
We collected data on

RCA2
Pure Pursuit
50 Hz
Velocity: 0.8


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df_physical = pd.read_csv('./PHYSICAL_RCA2_pp__2023-12-18_14-11-50_data.csv', comment='#')
df_sim = pd.read_csv('./SIM__2023-12-18_18-46-51_RCA2_pp_50Hz_vel0.8_noise_c[0.0, 0.0].csv', comment='#')

# Select the first 200 rows
df_physical = df_physical.head(500)
df_sim = df_sim.head(500)

state_keys = ['pose_x', 'pose_y', 'pose_theta', 'linear_vel_x', 'angular_vel_z', 'steering_angle']

# Plot each state key
for key in state_keys:
    plt.figure()  # Create a new figure
    plt.plot(df_physical[key], label='Physical')
    plt.plot(df_sim[key], label='Simulated')
    plt.title(key)
    plt.xlabel('Index')
    plt.ylabel(key)
    plt.legend()
    plt.show()
```
