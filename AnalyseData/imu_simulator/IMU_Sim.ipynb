```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('F1TENTH__2024-01-04_18-13-03Recording1_RCA2_neural_50Hz_vel0.9_noise_c[0.0, 0.0].csv', comment='#')
df = df.iloc[5:]

# Filter columns starting with 'imu'
columns_to_plot = ['imu_a_x', 'imu_a_y', 'imu_ang_z', 'steering_angle', 'slip_angle']


# Plot each of the imu columns
for col in columns_to_plot:
    plt.figure(figsize=(12, 6))
    plt.plot(df[col])
    plt.title(col)
    plt.show()
```

```python
# List of columns to plot
columns_to_plot = ['angular_vel_z', 'linear_vel_x', 'pose_theta', 'pose_x', 'pose_y', 'slip_angle']

# Plot a histogram for each of the specified columns
for col in columns_to_plot:
    plt.figure(figsize=(12, 6))
    plt.hist(df[col], bins=30, edgecolor='black')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```
