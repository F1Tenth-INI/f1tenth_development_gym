```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# NN
notebook_dir = os.getcwd()

df_mpc = pd.read_csv(notebook_dir+'/F1TENTH__2023-12-09_12-55-06_RCA1_mpc_50Hz.csv', comment='#')
df_nni = pd.read_csv(notebook_dir+'/F1TENTH__2023-12-09_12-57-34_RCA1_neural_50Hz.csv', comment='#')


plt.figure(figsize=(10, 8), dpi=150)
# plt.subplot(2, 1, 1)  # 2 rows, 1 column, first plot

plt.plot(df_mpc['translational_control_calculated'], color="red", label="MPC")
plt.plot(df_nni['translational_control_calculated'], color="blue", label="NNI")
plt.legend()  # Add legend
plt.ylabel('Speed command [m/s]')
plt.show()
plt.clf()

plt.figure(figsize=(10, 8), dpi=150)

# plt.subplot(2, 1, 2)  # 2 rows, 1 column, first plot

plt.plot(df_mpc['angular_control_calculated'], color="red", label="MPC")
plt.plot(df_nni['angular_control_calculated'], color="blue", label="NNI")
plt.legend()  # Add legend


plt.title('MPC vs Neural MPC imitator')
plt.xlabel('timestep')
plt.ylabel('Steering angle command [rad]')

plt.show()
```
