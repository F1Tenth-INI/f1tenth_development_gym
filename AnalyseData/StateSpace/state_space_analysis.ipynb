```python
import glob
import pandas as pd
import matplotlib.pyplot as plt

# Get a list of all CSV files in the 'data' directory
csv_files = glob.glob('./data/*.csv')

# List to hold dataframes
df_list = []

for file in csv_files:
    df = pd.read_csv(file, comment='#')
    df_list.append(df)

# Concatenate all dataframes in the list into a single dataframe
df = pd.concat(df_list, ignore_index=True)

```

```python
variables = ['linear_vel_x', 'angular_vel_z', 'angular_control_calculated', 'steering_angle', 'translational_control_calculated', 'slip_angle']

plt.figure(figsize=(10, 6))

for variable in variables:
    plt.figure(figsize=(10, 6))
    plt.hist(df[variable], bins=100, color='b', alpha=0.5, label=variable)  # Increased number of bins to 100
    plt.title(f'Histogram of {variable}')
    plt.xlabel(variable)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
variables = ['angular_vel_z', 'angular_control_calculated', 'steering_angle', 'angular_control_calculated', 'translational_control_calculated', 'slip_angle']
# colors = ['r', 'b', 'g', 'y']

plt.figure(figsize=(10, 6))

for variable in variables:
    plt.figure(figsize=(10, 6))
    plt.scatter(df['linear_vel_x'], df[variable], color="r", label=variable)
    plt.title(f'{variable} vs linear_vel_x')
    plt.xlabel('linear_vel_x')
    plt.ylabel(variable)
    plt.grid(True)
    plt.legend()
    plt.show()
```
