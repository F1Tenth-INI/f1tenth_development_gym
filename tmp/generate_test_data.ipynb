```python
import pandas as pd
import glob
import os
```

```python
dfs = []
# base_path = '../SI_Toolkit_ASF/Experiments/0_DNN/Recordings_preprocessed/Test'
base_path = '../SI_Toolkit_ASF/Experiments/Experiment-1/Recordings_preprocessed/Test'
for path in glob.glob(f'{base_path}/Trajectory-*.csv'):
    df = pd.read_csv(path, skiprows=8)
    df['filename'] = os.path.basename(path)
    dfs.append(df)
df = pd.concat(dfs)
```

```python
biggest_idx = df.groupby(['experiment_index', 'filename']).size().sort_values(ascending=False).index[0]
biggest_idx
```

```python
df_test = df[(df.experiment_index == biggest_idx[0]) & (df.filename == biggest_idx[1])]
df_test.describe()
df_test = df_test.set_index('time')
df_test
```

```python
# Copy header to new file
header_path = f'{base_path}/Trajectory-0.csv'
# out_file = f'../ExperimentRecordings/00_Test/ss_high_mu.csv'
out_file = f'../ExperimentRecordings/00_Test/ss_low_mu.csv'
with open(header_path, 'r') as f_header, open(out_file, 'w') as f_out:
    for line in f_header:
        if line[0] != '#':
            break
        f_out.write(line)
```

```python
df_test.to_csv(out_file, mode='a')
```
