```python
import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
```

```python
experiment_folder = 'SI_Toolkit_ASF/Experiments/0_AL'
dfs = {}
for i in range(5):
    df = pd.read_csv(f'{experiment_folder}/Models/Custom-11IN-ODE_module.py-STModel_high_mu-9OUT-0-{i}/log_training.csv', sep=';')
    lr = df.lr.iloc[0]
    dfs[lr] = df.mu
```

```python
for key, value in dfs.items():
    value.plot(legend=True, xlim=[0, 150], label=key)
plt.axhline(0.6, color='k', linestyle=':')
```
